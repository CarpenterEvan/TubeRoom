### Formatting the Dataframe with lambda 

Here I format the dataframe so it is more easily usable, I want each column to be a distinct parameter of the tube. 

I use unnamed lambda expressions to apply a function to every item in the dataframe, then split the dataframe by it's lists into multiple DataFrames, which I stitch together using pd.concat

In order to understand what the lambda function does to every item in the dataframe, I make two __named__ lambda expressions in the next cell

```python
first_lambda = lambda string: " ".join(string.split()) # This changes all separating space to just single space
second_lambda = lambda string: string.split(" ") # This splits strign into list with single space as delimiter

example_cell_in_dataframe = "This   makes a     list  that  properly  delimits by       space"

print(example_cell_in_dataframe)

first_modification = first_lambda(example_cell_in_dataframe)
print(first_modification)

second_modification = second_lambda(first_modification)
print(second_modification)
```

¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿¿

```python
print("Unmodified:")
display(df)

df = df.applymap(lambda string: " ".join(string.split()))

print("After first modification, all spaces are now 1 space:")
display(df)

df = df.applymap(lambda string: string.split(" "))

print("After second modification, all cells are a list delimited by 1 space:")
display(df)

# Now to split the dataframe into multiple dataframes by their lists!
ID = pd.DataFrame(df["ID"].tolist()[2:], columns=["tubeID",  "End",     "Received", "leakrate", "bend", "flagE"])
display(ID) # Only using one example for simplicity, but feel free to display more!
T1 = pd.DataFrame(df["T1"].tolist()[2:], columns=["T1Date", "Length", "Frequency",  "Tension",  "flag", "L"])
T2 = pd.DataFrame(df["T2"].tolist()[2:], columns=["dFrequency",  "dTension",   "dDays",    "flag2"])
DC = pd.DataFrame(df["DC"].tolist()[2:], columns=["DCday",   "sys",     "DC",   "HVseconds",     "DCflag"])
FV = pd.DataFrame(df["FV"].tolist()[2:], columns=["ENDday",  "done?",   "ok",       "Comment",   "None", "None"])

newdf = pd.concat((ID, T1, T2, DC, FV), axis=1)

print("Final Dataframe!!")
display(newdf)
```

# locate_tube_row(input_tubeID)
-    Takes in a tube's ID (MSUXXXXX where X is 1-9, in regex this is MSU[0-9]{5}) and finds the ID in the tubeID column of the dataframe. 
-    There are three default letters that I use for quick testing which translate to IDs: d, f, b.
-    If something not of the form MSU[0-9]{5} (MSU followed by 5 numbers) is entered, the defaults are checked
-    If the ID is not found, the row returned is -1

```python
DB = newdf
defaults = {"d": "MSU05123", # default (good)
            "f": "MSU01341", # fails multiple tests
            "b": "MSU00229"} # multiple tests missing--> "-----"
tubeID = input("input_tubeID: ") # in the .py this is just input_tubeID
try:
    istube = match("MSU[0-9]{5}", tubeID) 
    assert istube # this will raise an AssertionError if the ID does not look like "MSU" followed by 5 numbers
    # "tuberow, " unpacks just the first value of the tuple into the variable tuberow
    tuberow, = DB["tubeID"].index[ DB["tubeID"].str.contains(tubeID) ] # index of the row with True, which is row with ID
    print(tuberow)
except AssertionError: # This means ID was not valid, possibly a default key
    if len(tubeID) == 1 and tubeID in defaults.keys():
            tubeID = defaults[tubeID] # Default
            tuberow, = DB["tubeID"].index[ DB["tubeID"].str.contains(tubeID) ] # index of the row with True, which is row with ID
            print(tuberow)
    else: 
        print(-1)
except ValueError: # This means the ID is valid but not in the DB
    print(-2) 
```

# filter_columns(tuberow) 

```python
fullrow = DB.iloc[tuberow]
row_information = {"tubeID"           :    fullrow.tubeID,

                   "Shipment_date"    :  fullrow.Received,

                   "Bend_flag"        :     fullrow.flagE,
                   "T1_date"          :    fullrow.T1Date,
                   "T1_length"        :    fullrow.Length,

                   "T1_tension"       :   fullrow.Tension,
                   "T1_flag"          :      fullrow.flag,


                   "T2_tension_delta" :  fullrow.dTension,
                   "T2_time_delta"    :     fullrow.dDays,
                   "T2_flag"          :     fullrow.flag2,

                   "DC_date"          :     fullrow.DCday,

                   "DC_DC"            :        fullrow.DC,
                   "DC_seconds"       : fullrow.HVseconds,
                   "DC_flag"          :    fullrow.DCflag,


                   "Final_flag"       :        fullrow.ok}
print(row_information)
```
